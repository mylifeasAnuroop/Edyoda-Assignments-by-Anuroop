```python
import random
import sqlite3
22
# Define a class to represent food 
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = random.randint(1000, 9999)  # Generate a random 4-digit FoodID
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

#Class to represent users
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

#Main Food Ordering Application class
class FoodOrderingApp:
    def __init__(self):
        self.users = []
        self.admin_password = "admin123"
        self.current_user = None
        self.menu = []

#Add a new food item to the menu
    def add_food_item(self, name, quantity, price, discount, stock):
        food_item = FoodItem(name, quantity, price, discount, stock)
        self.menu.append(food_item)
        return food_item.food_id

#Edit an existing food item using its FoodID
    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for food_item in self.menu:
            if food_item.food_id == food_id:
                food_item.name = name
                food_item.quantity = quantity
                food_item.price = price
                food_item.discount = discount
                food_item.stock = stock
                return True
        return False

#View the list of all food items in the menu
    def view_all_food_items(self):
        return self.menu

#Remove a food item from the menu using its FoodID
    def remove_food_item(self, food_id):
        for food_item in self.menu:
            if food_item.food_id == food_id:
                self.menu.remove(food_item)
                return True
        return False

#Register a new user on the application
    def register_user(self, full_name, phone_number, email, address, password):
        user = User(full_name, phone_number, email, address, password)
        self.users.append(user)
        print("Registration successful!")

#Login a user into the application
    def login_user(self, email, password):
        user = next((u for u in self.users if u.email == email and u.password == password), None)
        if user:
            self.current_user = user
            print("Login successful!")
        else:
            print("Invalid email or password!")

#Place a new order
    def place_order(self):
        print("\n------ Place New Order ------")
        print("Menu:")
        for i, food_item in enumerate(self.menu, 1):
            print(f"{i}. {food_item.name} ({food_item.quantity}) [INR {food_item.price}]")

        selected_items = list(map(int, input("Enter item numbers separated by spaces: ").split()))
        order_items = [self.menu[item_number - 1] for item_number in selected_items]
        total_amount = sum(item.price for item in order_items)

        print("\nSelected Items:")
        for item in order_items:
            print(f"{item.name} ({item.quantity}) [INR {item.price}]")

        print(f"Total Amount: INR {total_amount}")

        confirm_order = input("Do you want to place this order? (yes/no): ").lower()
        if confirm_order == "yes":
            self.current_user.orders.append((order_items, total_amount))
            print("Order placed successfully!")
        else:
            print("Order cancelled.")

#View order history for the current user
    def order_history(self):
        print("\n------ Order History ------")
        for i, (order_items, total_amount) in enumerate(self.current_user.orders, 1):
            print(f"Order {i}:")
            for item in order_items:
                print(f"{item.name} ({item.quantity}) [INR {item.price}]")
            print(f"Total Amount: INR {total_amount}")
            print("--------------------------")

#Update profile for the current user
    def update_profile(self):
        print("\n------ Update Profile ------")
        self.current_user.full_name = input("Full Name: ")
        self.current_user.phone_number = input("Phone Number: ")
        self.current_user.email = input("Email: ")
        self.current_user.address = input("Address: ")
        self.current_user.password = input("Password: ")
        print("Profile updated successfully!!!")

#Define the main function to run the application
def main():
    food_app = FoodOrderingApp()

    while True:
        print("\n------ Food Ordering App ------")
        print("1. Admin Login")
        print("2. User Register")
        print("3. User Login")
        print("4. Exit")
        
# ADMIN PASSWORD IS = "admin123"

        choice = int(input("Enter your choice: "))

        if choice == 1:
            print("\n------ Admin Login ------")
            admin_password = input("Enter admin password: ")

            if admin_password == food_app.admin_password:
                while True:
                    print("\n------ Admin Menu ------")
                    print("1. Add New Food Item")
                    print("2. Edit Food Item")
                    print("3. View All Food Items")
                    print("4. Remove Food Item")
                    print("5. Logout")

                    admin_choice = int(input("Enter your choice: "))

                    if admin_choice == 1:
                        print("\n------ Add New Food Item ------")
                        name = input("Name: ")
                        quantity = input("Quantity: ")
                        price = int(input("Price: "))
                        discount = int(input("Discount: "))
                        stock = int(input("Stock: "))

                        food_id = food_app.add_food_item(name, quantity, price, discount, stock)
                        print(f"Food Item added with FoodID: {food_id}")

                    elif admin_choice == 2:
                        print("\n------ Edit Food Item ------")
                        food_id = int(input("Enter FoodID to edit: "))
                        name = input("Name: ")
                        quantity = input("Quantity: ")
                        price = int(input("Price: "))
                        discount = int(input("Discount: "))
                        stock = int(input("Stock: "))

                        if food_app.edit_food_item(food_id, name, quantity, price, discount, stock):
                            print("Food Item updated successfully!")
                        else:
                            print("FoodID not found!")

                    elif admin_choice == 3:
                        print("\n------ View All Food Items ------")
                        all_food_items = food_app.view_all_food_items()
                        for food_item in all_food_items:
                            print(f"FoodID: {food_item.food_id}, Name: {food_item.name}, Quantity: {food_item.quantity}, Price: {food_item.price}, Discount: {food_item.discount}, Stock: {food_item.stock}")

                    elif admin_choice == 4:
                        print("\n------ Remove Food Item ------")
                        food_id = int(input("Enter FoodID to remove: "))

                        if food_app.remove_food_item(food_id):
                            print("Food Item removed successfully!")
                        else:
                            print("FoodID not found!")

                    elif admin_choice == 5:
                        print("Logged out.")
                        break

                    else:
                        print("Invalid choice!")

            else:
                print("Invalid admin password!")

        elif choice == 2:
            print("\n------ User Registration ------")
            full_name = input("Full Name: ")
            phone_number = input("Phone Number: ")
            email = input("Email: ")
            address = input("Address: ")
            password = input("Password: ")
            food_app.register_user(full_name, phone_number, email, address, password)

        elif choice == 3:
            print("\n------ User Login ------")
            email = input("Email: ")
            password = input("Password: ")
            food_app.login_user(email, password)

            if food_app.current_user:
                while True:
                    print("\n------ User Menu ------")
                    print("1. Place New Order")
                    print("2. Order History")
                    print("3. Update Profile")
                    print("4. Logout")

                    user_choice = int(input("Enter your choice: "))

                    if user_choice == 1:
                        food_app.place_order()

                    elif user_choice == 2:
                        food_app.order_history()

                    elif user_choice == 3:
                        food_app.update_profile()

                    elif user_choice == 4:
                        food_app.current_user = None
                        print("Logged out.")
                        break

                    else:
                        print("Invalid choice!")

        elif choice == 4:
            print("Exiting the application.")
            break

        else:
            print("Invalid choice!")

#Call the main function to start the application
if __name__ == "__main__":
    main()
```
